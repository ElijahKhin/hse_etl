```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/khin/.cache/kagglehub/datasets/atulanandjha/temperature-readings-iot-devices/versions/1/IOT-temp.csv')
```

```python
# Поле noted_date переведите в формат ‘yyyy-MM-dd’ с типом данных date

df['noted_date'] = pd.to_datetime(df['noted_date'], dayfirst=1).dt.date
df['noted_date'] = pd.to_datetime(df['noted_date'])
```

```python
# Вычислите 5 самых жарких и самых холодных дней за весну

df[df.noted_date.dt.month.isin([3,4,5])]

# ... нет весенних месяцев
```

```python
# Отфильтруйте out/in = in

df[df['out/in'] == 'In']
```

```python
# Очистите температуру по 5-му и 95-му процентилю

# Вычисляем 5-й и 95-й процентили
lower_bound = df['temp'].quantile(0.05)
upper_bound = df['temp'].quantile(0.95)

# Фильтруем данные
df_filtered = df[(df['temp'] >= lower_bound) & (df['temp'] <= upper_bound)]
```
